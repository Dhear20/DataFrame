```python
import pandas as pd

#Baca file Excel (import)
df_excel = pd.read_excel('jml_sampah.xlsx', sheet_name='data')
df_excel
```

```python
df_sampah = pd.DataFrame(df_excel, columns=['nama_kabupaten_kota','jumlah_produksi_sampah','tahun'])
df_sampah

df_sampah.to_csv('data_sampah.csv')
df_sampah.to_excel('data_sampah.xlsx')
```

```python
import numpy as np

d23 ={} #jumlah sampah pada tahun 2015

for i , row in df_excel.iterrows():
    tahun = row['tahun']
    sampah = row['jumlah_produksi_sampah']

    if tahun == 2023 not in d23 :
        d23[tahun] = sampah
    elif tahun == 2023 in d23 :
        d23[tahun] += sampah
print  ('=============== Data Tahun 2015 ===============')
for i,j in d23.items():
    print(f"Jumlah sampah pada tahun {i} adalah {j:.2f} ton")
    
df_t2023 = pd.DataFrame ([
    {'Tahun':i,'Total/tahun':j}
    for i,j in d23.items()
])

df_t2023

df_t2023.to_csv('df_t2023.csv')
df_t2023.to_excel('df_t2023.xlsx')
```

```python
thn = {}

for i,j in df_excel.iterrows():
     tahun = j['tahun']
     tsampah = j['jumlah_produksi_sampah']
     
     if tahun not in thn:
          thn[tahun] = tsampah 
     else :
          thn[tahun] += tsampah

print  ('=============== Data per tahun ==================')         
for i,j in thn.items():
     print(f'Pada tahun {i} terdapat {j:.2f} ton sampah')

df_exceltahun = pd.DataFrame([
   {'Tahun':i,'Total Sampah':j}
   for i,j in thn.items()
])

df_exceltahun

df_exceltahun.to_csv('df_sampahtahun.csv')
df_exceltahun.to_excel('df_sampahtahun.xlsx')
```

```python
kab_kota_thn = {}

for i,j in df_excel.iterrows():
     daerah = j['nama_kabupaten_kota']
     sampah = j['jumlah_produksi_sampah']
     
     if daerah not in kab_kota_thn : 
          kab_kota_thn[daerah] = sampah
     else : 
          kab_kota_thn[daerah] += sampah
print  ('=============== Data per Daerah ===============')
for i,j in kab_kota_thn.items():
     print(f'Pada daerah {i} terdapat {j:.2f} ton sampah')
     
df_sampahdaerah = pd.DataFrame([
     {'Nama Kabupaten/Kota':i,'Total Sampah':j}
     for i,j in kab_kota_thn.items()
])
df_sampahdaerah

df_sampahdaerah.to_csv('df_sampahkabkota.csv')
df_sampahdaerah.to_excel('df_sampahkabkota.xlsx')
```
